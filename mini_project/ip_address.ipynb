```python
with open ("example_log.txt",mode='r') as df:
    lst=[]
    for i in df :
        lst.append(i)
lst 
```

```python
lst2=[]
for i in lst:
    lst2.append(i.split()[0])
len(lst2)
```

```python
a=list(filter(lambda x:x>="50",lst2))
len(a)
```
