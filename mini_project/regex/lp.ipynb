* Regex(Regular Expression) makes it easy to extract numeric values and units (GB, TB) in one step — flexible, fast, and perfect for mixed storage strings.

```python
df['storage'].unique()
```

```python
import re
def convert_to_gb(value):
    parts = re.findall(r'(\d+\.?\d*)\s*(TB|GB)', str(value))
    total_gb = 0
    for amount, unit in parts:
        amount = float(amount)
        if unit.upper() == 'TB':
            amount *= 1024  # Convert TB → GB
        total_gb += amount
    return total_gb
```

```python
df['storage_gb'] = df['storage'].apply(convert_to_gb).astype(int)    # Apply the conversion function and convert to integer
df.head(1)
```

### Missing values :
* Missing At Random
* Missing Not At Random
* Missing Completely At Random


| Type of Missingness                 | Dependency                         | Example                      | Treatment Methods                                  |
| ----------------------------------- | ---------------------------------- | ---------------------------- | -------------------------------------------------- |
| Missing Completely at Random (MCAR) | Independent of all data            | Random system failure        | Mean/Median imputation, Deletion                   |
| Missing at Random (MAR)             | Depends on observed variables      | Education missing due to Age | Regression / KNN / Multiple Imputation             |
| Missing Not at Random (MNAR)        | Depends on missing variable itself | High salary not disclosed    | Domain inference, Missing indicator, External data |

| Method           | Description                                  | Best Used For        | Example                              |
| ---------------- | -------------------------------------------- | -------------------- | ------------------------------------ |
| `fillna()`       | Fills NaN with constant/statistical value    | General missing data | `df.fillna(0)`                       |
| `ffill()`        | Fills NaN with previous valid value          | Sequential/time data | `df.ffill()`                         |
| `bfill()`        | Fills NaN with next valid value              | Sequential/time data | `df.bfill()`                         |
| `interpolate()`  | Estimates missing values using interpolation | Numeric sequences    | `df.interpolate()`                   |
| Mean/Median/Mode | Statistical imputation                       | Numeric/categorical  | `df['col'].fillna(df['col'].mean())` |
| Model-based      | Predicts missing values                      | Complex dependencies | `KNNImputer`, `Regression`            |
