### Boolean Masking


```python
n1=np.arange(2,20)
```

```python
condition=n1>7
n1[condition]
```

```python
n1[(n1>7) & (n1<15)]
```

```python
n1[(n1>7) | (n1<15)]
```

### Linear Equation

```python
A = np.asarray([[30, -1],               
                [50, -1]])          # asarray = array

Inv_A = np.linalg.inv(A)  # linalg = linear algebra
print(Inv_A)
    
```

```python
B=np.asarray([-1000,-100])
```

```python
X=np.dot(Inv_A,B)
X
```

### Stacking

```python
n1=np.arange(1,3)
n2=np.arange(3,5)
```

```python
np.hstack((n1,n2))
```

```python
np.vstack((n1,n2))
```

```python
np.concatenate((n1,n2),axis=0)
```

```python
np.concatenate((n1.reshape(-1,1),n2.reshape(-1,1)),axis=1)
```
