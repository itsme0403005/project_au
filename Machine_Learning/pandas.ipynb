loc and iloc function


```python
df.iloc[0::2,0::2]  # start_row:end_row:row_step, start_col:end_col:col_step
```

```python
df.loc[0:2,"Name"::2]     # row name start: row name end : step, column name start : column name end : step
```

```python
df.loc[0:2,['Name','Subjects']]     # row name start: row name end : step, column name start : column name end : step
```

```python
# df['Marks']==100 
# df[df['Marks']==100]
```

```python
# x.loc[:,"Marks"]=125
df.loc[df['Marks']==100,'Marks']=125
```

```python
df
```

```python
df.loc[df['Subjects']=="Maths",'Subjects']="Physics"
df
```

```python
df.loc[df['Marks']>100,'Marks']=df['Marks']+100
df
```

```python
df.loc[(df['Marks']>250) & (df['Marks']<500),['Marks']]=df['Marks']-100
df
```
